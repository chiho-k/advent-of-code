```python
import csv
```

```python
file = open("calibration value.txt", "r")
```

```python
firstNumbers = []
lastNumbers = []
fileData = []
for line in file:
    fileData.append(line)
```

```python
def calibration(list):
    combinedNumbersList = []
    combinedNumbers = 0
    for line in list:
        print(line)
        combinedNumber = 0
        for letter in line:
            try:
                firstNumber = int(letter)
                firstNumbers.append(firstNumber)
                combinedNumber += firstNumber*10
                break
            except:
                continue
        reversedLine = line[::-1]
        for letter in reversedLine:
            try:
                lastNumber = int(letter)
                lastNumbers.append(lastNumber)
                combinedNumber += lastNumber
                break
            except:
                continue
        combinedNumbersList.append(combinedNumber)
        print(combinedNumber)
        #print(combinedNumbersList)
        combinedNumbers += combinedNumber
        print(combinedNumbers)
    return combinedNumbers, combinedNumbersList
#print(fileData)
calibrated = calibration(fileData)
print(calibrated)
```

```python
numberWordsFile = open("number words.csv", "r")
numberWordsDict = csv.DictReader(numberWordsFile)
numberWords = []
for row in numberWordsDict:
    numberWords.append(row)
    print(row)
```

```python
words = []
for row in numberWords:
    words.append(row["word"])
```

```python
def wordConverter(list):
    newLines = []
    for line in list:
        newLine = line
        for i in range(0, len(newLine)):
            for word in words:
                for row in numberWords:
                    if row["word"] == word:
                        number = row["number"]
                if word in newLine[i:i+len(word)]:
                    newLine = newLine[:i] + number + newLine[i+len(word)-1:]
            else:
                continue
        newLines.append(newLine)
    return newLines
```

```python
newData = wordConverter(fileData)
calibratedNew = calibration(newData)
print(calibratedNew)
```

```python
comparison = []
differenceTotal = 0
for n in range(0,1000):
    if calibrated[1][n] != calibratedNew[1][n]:
        difference = calibrated[1][n] - calibratedNew[1][n]
        differenceTotal += difference
        comparison.append((calibrated[1][n], calibratedNew[1][n], fileData[n], difference, differenceTotal))
    else:
        continue
print(comparison)
print(differenceTotal)
```
